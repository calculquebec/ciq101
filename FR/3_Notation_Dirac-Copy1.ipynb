# Notation de Dirac

La notation de Dirac (bra ket) est une façon pratique de décrire des [vecteurs](https://fr.wikipedia.org/wiki/Vecteur).
* ket ∣ ⟩ est un vecteur d’état représenté par une colonne
* bra ⟨ ∣ est un vecteur d’état [conjugué](https://fr.wikipedia.org/wiki/Conjugu%C3%A9) représenté par une rangée

**État de superposition $\psi$** <br>
 <br>
* Le Conjugué (*) inverse le signe (nombre complex i devient -i)
* Obèle (†) signifie une transposition entre rangée et colonne

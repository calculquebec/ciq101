# Portes logiques quantiques

Une porte logique quantique est une **matrice unitaire** (chaque colonne et rangée est [orthonormal](https://en.wikipedia.org/wiki/Orthonormality#:~:text=Orthogonal%20means%20that%20the%20vectors,is%20called%20an%20orthonormal%20basis.)) :
* La matrice est carrée (même nombre de rangées que de colonnes)
* Ordre de la matrice est égale à $2^n$ x $2^n$
* La norme de la matrice est de 1
* L'inverse de la matrice est égale à la transposition conjuguée: $U^* = U^{-1}$
* Par conséquent, toutes les opérations sur les qubits sont réversibles

Comment exprimer mathématiquement un circuit quantique :

État de Bell:
* État qui ne peut pas être décomposé comme un produit tensoriel d'états de qubits simples
* Pour deux qubits, les états $\ket{01}$ et $\ket{10}$ ont une probabilité de zéro
* Les états $\ket{00}$ et $\ket{11}$ ont une probabilité de 50%
* Les états de Bell sont des exemples d'intrication qui sont utilisés pour la cryptographie quantique et la correction d'erreur sur les qubits

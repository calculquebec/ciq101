# Qubits

## Sphère de Bloch 

* **Axes rotationnels (impulsions contrôles la rotation d'un qubit autour des axes)**
  * $\ket{0}$, $\ket{1}$ (base computationnelle, Pauli-Z)
  * $\ket{+}$, $\ket{-}$ (base de Hadamard, Pauli-X)
  * $\ket{i}$, $\ket{-i}$ (base imaginaire, Pauli-Y)

## État de superposition d'un qubit

* $\ket{\psi} = \alpha\ket{0} + \beta\ket{1} = \cos(\frac{\theta}{2})\ket{0} + e^{i\phi}\sin(\frac{\theta}{2})\ket{1}$ $\rightarrow$  [détails sur la transformation en coordonnées polaires](https://medium.com/quantum-untangled/quantum-states-and-the-bloch-sphere-9f3c0c445ea3)

* Généralement, on fait une mesure le long de l'axe de la base computationnelle
  * La mesure réduit l'état de superposition $\ket{\psi}$ à l'un des états de la base computationnelle, $\ket{0}$ ou $\ket{1}$, avec une probabilité correspondante au carré des coefficients $\alpha$ ou $\beta$
<br>
<br>
* **Probabilités**: $|\alpha|^2 + |\beta|^2 = 1$, la somme des probabilités est toujours 1.
  * Norm (length): $|\psi| = \sqrt{\alpha^2 + \beta^2} = 1$

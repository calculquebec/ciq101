# Qubits

## Bloch sphere

* **Rotational axes (pulses control the rotation of a qubit about the axes)**
  * $\ket{0}$, $\ket{1}$ (computational basis, Pauli-Z)
  * $\ket{+}$, $\ket{-}$ (Hadamard basis, Pauli-X)
  * $\ket{i}$, $\ket{-i}$ (imaginary basis, Pauli-Y)

## Qubit superposition of states

* $\ket{\psi} = \alpha\ket{0} + \beta\ket{1} = \cos(\frac{\theta}{2})\ket{0} + e^{i\phi}\sin(\frac{\theta}{2})\ket{1}$ $\rightarrow$  [transformation into polar coordinates](https://medium.com/quantum-untangled/quantum-states-and-the-bloch-sphere-9f3c0c445ea3)

* In general, a measurement is performed along the computational basis
  * A measurement collapses the superposition state $\ket{\psi}$ into one of the states of the computational basis, $\ket{0}$ ou $\ket{1}$, with a probability of $\alpha^2$ or $\beta^2$
<br>
<br>
* **Probabilities**: $|\alpha|^2 + |\beta|^2 = 1$, the sum of probabilities is always 1.
  * Norm (length): $|\psi| = \sqrt{\alpha^2 + \beta^2} = 1$

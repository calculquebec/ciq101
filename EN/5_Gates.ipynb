# Quantum logic gates

A quantum logic gate is a **[unitary matrix](https://mathworld.wolfram.com/UnitaryMatrix.html)** (each column and row is: [orthonormal](https://en.wikipedia.org/wiki/Orthonormality#:~:text=Orthogonal%20means%20that%20the%20vectors,is%20called%20an%20orthonormal%20basis.)):
* The matrix is square (same number of rows as columns)
* The order of the matrix is $2^n$ x $2^n$
* The [norm of the matrix](https://mathworld.wolfram.com/MatrixNorm.html) is 1
* The inverse of the matrix is equal to the conjugate transpose: $U^* = U^{-1}$
* As a result, all operations on qubits are reversible.

How to express a quantum circuit mathematically:

Bell state:
* State that cannot be decomposed as a tensor product of single qubit states
* The probability is zero for the $\ket{01}$ and $\ket{10}$ states
* The $\ket{00}$ and $\ket{11}$ have a probability of 50%
* Bell states are examples of entanglement that are used for quantum cryptography and error correction on qubits.
